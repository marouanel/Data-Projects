# FIFA World Cup Tweets

In this kernel, I will perform a preliminary analysis through a dataset containing a random collection of tweets starting from the Round of 16 till the World Cup Finals, & explore some of the Natural Language Processing techniques in Python to understand the sentiments behind tweets The dataset is available on [Kaggle]((https://www.kaggle.com/rgupta09/world-cup-2018-tweets)

### Multiple Data sources

To get the IDs and other geo information about the countries, I got the csv file from the following git [https://github.com/lukes/ISO-3166-Countries-with-Regional-Codes/tree/master/all]

To get information about game days and the group stages, I downloaded a CSV file from the following website [https://fixturedownload.com/results/fifa-world-cup-2018]




```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer


import matplotlib.pyplot as plt
import seaborn as sns
import json 
import pandas as pd 

from geotext import GeoText
import re
from pandas.io.json import json_normalize #package for flattening json in pandas df


%matplotlib inline
sns.set(style="darkgrid")
```

```python
Tweets = pd.read_csv("data/FIFA.csv",encoding = "ISO-8859-1")
geo_tweets = Tweets.dropna(subset=['Place']).reset_index()
geo_tweets['Date'] = pd.to_datetime(geo_tweets['Date'])
geo_tweets.sort_values('Date',inplace=True)
print("Number of Tweets: ", Tweets.shape[0])
geo_tweets.head()
```

# Adding coutries to tweets 

You can see that places are not really intuitive to get, that is why we need a more standard description of the location.

I am using GeoText API to extract the countries from the 'Place' column, and I am matching the exctracted code with the second dataset containing inforation about countries 

PS: I noticed that England and Korea were not picked up by GeoText so I changed them in the dataframe ^^ 


```python
f_countries = lambda x: str(GeoText(x).country_mentions.popitem()[0]).lower() if (GeoText(x).country_mentions) else 'nada'
f_day = lambda x: x.day
f_lower = lambda x : str(x).lower()
geo_tweets['day'] = geo_tweets['Date'].apply(f_day)

geo_tweets['Place'] = geo_tweets['Place'].str.replace(r'(^.*england.*$)', 'great britain england ', flags=re.IGNORECASE)
geo_tweets['Place'] = geo_tweets['Place'].str.replace(r'(^.*korea.*$)', 'South Korea', flags=re.IGNORECASE)

geo_tweets['country'] = geo_tweets['Place'].apply(f_countries)
iso2_codes = pd.read_csv('data/iso2.csv')


iso2_codes['country'] = iso2_codes['alpha-2'].apply(f_lower)
iso2_codes.columns = ['country_name', 'alpha-2', 'alpha-3', 'country-code', 'iso_3166-2', 'region',
       'sub-region', 'intermediate-region', 'region-code', 'sub-region-code',
       'intermediate-region-code', 'country']
geo_tweets = pd.merge(geo_tweets,iso2_codes,on='country', how='left')
print("Number of Tweets: ", geo_tweets.shape[0])
geo_tweets.head()
```

# Getting Overlap of Games and Tweets

I am loading the game table and filtering it to only get the games that happened in the same time frame as the tweets (so basically getting rid of the group stage)


```python
#load json object
import re
with open('data/data.json') as f:
    d = json.load(f)

teams = json_normalize(d['teams'])
teams['iso2'] = teams['iso2'].str.replace(r'(^.*gb-eng.*$)', 'gb', flags=re.IGNORECASE)
games = pd.read_csv('data/games_table.csv', parse_dates=[1], dayfirst=True)

games['Home Team'] = games['Home Team'].str.replace(r'(^.*korea.*$)', 'South Korea', flags=re.IGNORECASE)
games['Home Team'] = games['Home Team'].str.replace(r'(^.*england.*$)', 'united kingdom britain', flags=re.IGNORECASE)
game_dates = games[games['Date'] > geo_tweets.Date.min()]
game_dates = game_dates[game_dates['Date'] < geo_tweets.Date.max()]

game_dates['color'], rounds = pd.factorize(game_dates['Round Number'])
geo_tweet_teams = pd.merge(geo_tweets,teams,left_on='country', right_on='iso2', how='left').dropna(subset=['iso2'],axis=0)


game_dates
```

# Exploring Topics discussed during the word cup

To get a sense of what Each country talked about, I am using the Author Topic Model to link between countries, topics and tweets

Before doing so I am removing the retweets (because content wise they are basicaly duplicates)

The pre-processing steps follow the general guidelines of working with text in the Nature Language Processing field and the reviewed work on service discovery RS:

####  Tokenization: dividing a stream of text into a sequence of words according to a delimiter (space in this case)
#### Pruning: Removing words that do not add meaning to the sentence, also known as stop-words removal, a list of stopwords was taken from the Natural Language Toolkit (nltk) python library to this end
##### item Stemming: Stripping words from suffixes and/or prefixes to get to the word stem/root. This way word like 'map','mapping' and 'maps' would be counted as 'map' adding relevance for the model. The Porter Stemmer of the nltk library was used.



```python
geo_tweet_teams = geo_tweet_teams.drop_duplicates(subset = ['Tweet']).reset_index()
```

```python
geo_tweet_teams.shape[0]
```

```python
#storing the mashup corpus as a list of words
corpus_list=[]
#storing the corpus as strings
corpus_string = []


for i in range(0, geo_tweet_teams.shape[0]):
    
  
    
    #Extract Description and tags while removing special characters 
    review = str(geo_tweet_teams['Tweet'][i])
    review = review.lower()  
    review = review.split()
    #Changing each word to stem form
    ps = PorterStemmer()
    stopwords_ = list(set(stopwords.words('english')))
    review = [ps.stem(word) for word in review if not word in set(stopwords_)]
    corpus_list.append(review)
    review = ' '.join(review)
    corpus_string.append(review)
    

print('Example Tweet \n',geo_tweet_teams['Tweet'][0],'\n After Pre-processing')
corpus_string[0]
```

### Creating a dictionay linking each country to its tweets

```python
unique_countries = geo_tweet_teams['country_name'].unique()
country2tweets = dict()
for c in unique_countries:
    country2tweets[c] = geo_tweet_teams.loc[geo_tweet_teams.country_name ==c].index
    

```

```python
from gensim.corpora import Dictionary
dictionary = Dictionary(corpus_list)
max_freq = 0.5
min_wordcount = 20
dictionary.filter_extremes(no_below=min_wordcount, no_above=max_freq)

dictionary[0] 
corpus = [dictionary.doc2bow(doc) for doc in corpus_list]
```

# What are the countries that tweet about similar topics ?

To get a visual representation of the countries, topics and tweets. To do so, we take all the author-topic distributions (stored in model.state.gamma) and embed them in a 2D space. To do this, we reduce the dimensionality of this data using t-SNE 

I color coded the result with each team's corresponding group at the group stage 

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0)
smallest_author = 0  # Ignore authors with documents less than this.
authors = [model.author2id[a] for a in model.author2id.keys() if len(model.author2doc[a]) > smallest_author]
_ = tsne.fit_transform(model.state.gamma[authors, :])  # Result stored in tsne.embedding_
author_names = [model.id2author[a] for a in authors]

gp_groups = pd.DataFrame(games[:60].groupby(['Home Team'])['Group'].unique()).reset_index()
gp_groups.columns=['team','group']
gp_groups['team'].replace('united kingdom britain','United Kingdom of Great Britain and Northern Ireland',inplace=True)
gp_groups['team'].replace('Russia','Russian Federation',inplace=True)
gp_groups['team'].replace('Iran','Iran (Islamic Republic of)',inplace=True)
gp_groups['team'].replace('South Korea','Korea (Republic of)',inplace=True)
gp_groups['country'] = gp_groups['team'].apply(f_countries)
f_string = lambda x: x[0]
gp_groups['group'] = gp_groups['group'].apply(f_string)
author_group =[]
for each in author_names:
    g =  gp_groups.loc[gp_groups['team'] == (each)]
    author_group.append((g['group'].values))

from bokeh.models import HoverTool
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.transform import factor_cmap
from bokeh import palettes 
from bokeh.io import output_notebook
output_notebook()

x = tsne.embedding_[:, 0]
y = tsne.embedding_[:, 1]
author_names = [model.id2author[a] for a in authors]

# Radius of each point corresponds to the number of documents attributed to that author.
#scale = 0.01
author_sizes = [len(model.author2doc[a]) for a in author_names]
#radii = [size * scale for size in author_sizes]
radii = np.ones(len(author_names))*7
source = ColumnDataSource(
        data=dict(
            x=x,
            y=y,
            author_names=author_names,
            author_sizes=author_sizes,
            group=author_group,
            radii=radii,
        )
    )

# Add author names and sizes to mouse-over info.
hover = HoverTool(
        tooltips=[
        ("Team", "@author_names"),
        ("Tweets", "@author_sizes"),
        ("Group", "@group"),
        ]
    )
color_mapper = factor_cmap('group',factors=['Group A', 'Group B','Group C', 'Group D', 'Group E', 'Group F', 'Group G', 'Group H'], 
                                      palette=palettes.Category10[8])


p = figure(tools=[hover, 'crosshair,pan,wheel_zoom,box_zoom,reset,save,lasso_select'])
p.scatter('x', 'y', radius='radii', source=source, fill_alpha=0.6, line_color = None,fill_color = color_mapper)
show(p)
```
